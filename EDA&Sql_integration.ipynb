```python

import pandas as pd
from dotenv import load_dotenv
import os
```

```python
df_route_1  = pd.read_csv('Bus_details/AP_Buses.csv')
df_route_2  = pd.read_csv('Bus_details/Assam_Buses.csv')
df_route_3  = pd.read_csv('Bus_details/Chandigarh_Buses.csv')
df_route_4  = pd.read_csv('Bus_details/Goa_Buses.csv')
df_route_5  = pd.read_csv('Bus_details/Himachal_Buses.csv')
df_route_6  = pd.read_csv('Bus_details/Jk_Buses.csv')
df_route_7  = pd.read_csv('Bus_details/Kerala_Buses.csv')
df_route_8  = pd.read_csv('Bus_details/Rajasthan_Buses.csv')
df_route_9  = pd.read_csv('Bus_details/TLbuses.csv')
df_route_10 = pd.read_csv('Bus_details/UP_Buses.csv')

df_Route_details=pd.concat([df_route_1,df_route_2,df_route_3,df_route_4,df_route_5,
                            df_route_6,df_route_7,df_route_8,df_route_9,df_route_10],ignore_index=True)
df_Route_details
```

```python
df_Route_details["Price"] =  df_Route_details["Price"].str.replace("INR"," ")
df_Route_details["Price"] = df_Route_details["Price"].astype(float)
df_Route_details["Price"].dtype
```

```python
df_Route_details.isna()
```

```python
import numpy as np
df_Route_details["Ratings"] = df_Route_details["Ratings"].fillna(0)
df_Route_details        =       df_Route_details.replace({np.nan : None})
```

```python
df_Route_details['Price'] = df_Route_details['Price'].astype(str).str.replace('[^0-9.]', '', regex=True)
print(df_Route_details['Price'].unique())
```

```python
df_Route_details['Price'] = df_Route_details['Price'].replace('', np.nan)
df_Route_details.dropna(subset=['Price'], inplace=True) # Example: Remove rows with NaN price
```

```python
df_Route_details = df_Route_details.dropna()
```

```python
df_Route_details.info()
```

```python
df_Route_details['Price'] = df_Route_details['Price'].astype(str).str.replace('[^0-9.]', '', regex=True)
df_Route_details['Price'] = pd.to_numeric(df_Route_details['Price'], errors='coerce')
df_Route_details.dropna(subset=['Price'], inplace=True)
```

```python
def convert_to_time(time_str):
    """
    Converts a time string to a datetime.time object, handling different formats and errors.
    """
    formats = ['%H:%M:%S', '%H:%M']  # Try both HH:MM:SS and HH:MM
    for fmt in formats:
        try:
            return pd.to_datetime(time_str, format=fmt).time()
        except ValueError:
            pass  # If one format fails, try the next
    return pd.NaT  # Return NaT if no format works

# Convert 'Start_time' to datetime.time objects
df_Route_details['Start_time'] = df_Route_details['Start_time'].apply(convert_to_time)

# Convert 'End_time' to datetime.time objects
df_Route_details['End_time'] = df_Route_details['End_time'].apply(convert_to_time)

print(df_Route_details[['Start_time', 'End_time']].head())
print("***"*50)
print(df_Route_details[['Start_time', 'End_time']].info())
print("***"*50)
```

```python
df_Route_details['Seats_Available'] = df_Route_details['Seats_Available'].astype(str).str.replace('[^0-9]', '', regex=True)
df_Route_details['Seats_Available'] = pd.to_numeric(df_Route_details['Seats_Available'], errors='coerce')
df_Route_details.dropna(subset=['Seats_Available'], inplace=True)
```

```python
import mysql.connector
from mysql.connector import Error
```

```python
load_dotenv()
DB_HOST = os.getenv("DB_HOST")
DB_USER = os.getenv("DB_USER")
DB_PASSWORD = os.getenv("DB_PASSWORD")
try:
    conn = mysql.connector.connect(
        host=DB_HOST,
        user=DB_USER,
        password=DB_PASSWORD
        )
    cursor = conn.cursor()
    cursor.execute("CREATE DATABASE IF NOT EXISTS RED_BUSES")
    
    # 4. commit on connection
    conn.commit()
    print("Database 'RED_BUSES' created successfully.")
except Error as err:
    print(f"Error: {err}")
    
finally:
    # 5. clean up
    if cursor:
        cursor.close()

```

```python
load_dotenv()
DB_HOST = os.getenv("DB_HOST")
DB_USER = os.getenv("DB_USER")
DB_PASSWORD = os.getenv("DB_PASSWORD")
try:
    conn = mysql.connector.connect(
        host=DB_HOST,
        user=DB_USER,
        password=DB_PASSWORD
        )
    cursor = conn.cursor()
    cursor.execute("USE RED_BUSES")
    cursor.execute('''CREATE TABLE IF NOT EXISTS bus_details(
                  ID INT AUTO_INCREMENT PRIMARY KEY,
                  Bus_name VARCHAR(255) NOT NULL,
                  Bus_type VARCHAR(255) NOT NULL,
                  Start_time VARCHAR(255) NOT NULL,
                  End_time VARCHAR(255) NOT NULL,
                  Total_duration VARCHAR(255) NOT NULL,
                  Price DECIMAL(10,2) NULL,
                  Seats_Available VARCHAR(255) NULL,
                  Ratings Float  NULL,
                  Route_link VARCHAR(255) NULL,
                  Route_name VARCHAR(255) NULL
                  )''')
    
    # 4. commit on connection
    conn.commit()
    print("Table Created successfully.")
except Error as err:
    print(f"Error: {err}")
    
finally:
    # 5. clean up
    if cursor:
        cursor.close()
```

```python
df_Route_details = pd.read_csv("Route_details.csv")
```

```python
insert_query = '''INSERT INTO bus_details(
                    Bus_name,
                    Bus_type,
                    Start_time,
                    End_time,
                    Total_duration,
                    Price,
                    Seats_Available,
                    Ratings,
                    Route_link,
                    Route_name)
                  VALUES(%s, %s, %s, %s, %s, %s, %s, %s, %s, %s)'''

data = df_Route_details[[
    'Bus_name', 'Bus_type', 'Start_time', 'End_time', 'Total_duration',
    'Price', 'Seats_Available', 'Ratings', 'Route_link', 'Route_name'
]].values.tolist()
```

```python
load_dotenv()
DB_HOST = os.getenv("DB_HOST")
DB_USER = os.getenv("DB_USER")
DB_PASSWORD = os.getenv("DB_PASSWORD")

try:
    conn = mysql.connector.connect(
        host=DB_HOST,
        user=DB_USER,
        password=DB_PASSWORD,
        )
    cursor = conn.cursor()
    cursor.execute("USE RED_BUSES")
    cursor.executemany(insert_query, data)
    conn.commit()
    print("Values inserted successfully")
except NotImplementedError as err:
    print(f"Error: {err}")
finally:
    if conn.is_connected():
        cursor.close()
        conn.close()
        print("MySQL connection is closed.")
```
